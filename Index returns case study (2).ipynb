```python
import pandas as pd
import numpy as np
import scipy as sp
from scipy import stats

 
df = pd.read_csv('sandp.csv')


df['Date'] =  pd.to_datetime(df['Date'])



l=[]
x=0
while x < 7561:
    timestamp_value = pd.Timestamp.isoweekday(df.iloc[x, df.columns.get_loc('Date')])
    l.append(timestamp_value)
    x += 1
df['Timestamp'] = l



op_minus_cl_friday=[]
op_minus_cl_monday=[]
y=0
while y < 7561:
    day = df.iloc[y, df.columns.get_loc('Timestamp')]
    op = df.iloc[y, df.columns.get_loc('Open')]
    cl = df.iloc[y, df.columns.get_loc('Close')]
    if day == 5:
        op_minus_cl_friday.append(((op-cl)/op))
    else:
        op_minus_cl_friday.append(0)
    if day == 1:
        op_minus_cl_monday.append(((op-cl)/op))
    else:
        op_minus_cl_monday.append(0)
    y += 1
    
df['Friday Return'] = op_minus_cl_friday
df['Monday Return'] = op_minus_cl_monday

correlation1=[]
correlation2=[]
z=0
while z < 7561:
    Friday_Return = df.iloc[z, df.columns.get_loc('Friday Return')]
    if Friday_Return != 0:
        correlation1.append(Friday_Return)
    z  += 1

i=0
while i < 7561:
    Monday_Return= df.iloc[i, df.columns.get_loc('Monday Return')]
    if Monday_Return != 0:
        correlation2.append(Monday_Return)
    i  += 1

df1= {'friday': correlation1}
df2= {'monday': correlation2}

df1= pd.DataFrame(df1)
df2= pd.DataFrame(df2)

df3= pd.concat([df1,df2], ignore_index=True, axis=1)

mean_friday = pd.DataFrame.mean(df3[0])
mean_monday = pd.DataFrame.mean(df3[1])
mean_friday_var = pd.DataFrame.var(df3[0])
mean_monday_var = pd.DataFrame.var(df3[1])


all_return=[]
a=0
while a < 7561:
    o = df.iloc[a, df.columns.get_loc('Open')]
    c = df.iloc[a, df.columns.get_loc('Close')]
    all_return.append(((o-c)/o))
    a += 1

df['All Return'] = all_return

all_return_mean = df['All Return'].mean()

return_correlation = df3[0].corr(df3[1])


#descriptive statistics
abar = df3[0].mean()
avar = df3[0].var(ddof=1)
na = df3[0].size
adof = na - 1

bbar = df3[1].mean()
bvar = df3[1].var(ddof=1)
nb = df3[1].size
bdof = nb - 1

# inferred statistics
t2, p2 = sp.stats.ttest_ind_from_stats(abar, np.sqrt(avar), na,
                              bbar, np.sqrt(bvar), nb,
                              equal_var=False)
print("T-test values: t = %g  p = %g" % (t2, p2))
print('correlation:', return_correlation)
```
