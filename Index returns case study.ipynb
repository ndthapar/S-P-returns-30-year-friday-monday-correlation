```python
import pandas as pd
import numpy as np
df = pd.read_csv('sandp.csv')


df['Date'] =  pd.to_datetime(df['Date'])



l=[]
x=0
while x < 7561:
    timestamp_value = pd.Timestamp.isoweekday(df.iloc[x, df.columns.get_loc('Date')])
    l.append(timestamp_value)
    x += 1
df['Timestamp'] = l



op_minus_cl_friday=[]
op_minus_cl_monday=[]
y=0
while y < 7561:
    day = df.iloc[y, df.columns.get_loc('Timestamp')]
    op = df.iloc[y, df.columns.get_loc('Open')]
    cl = df.iloc[y, df.columns.get_loc('Close')]
    if day == 5:
        op_minus_cl_friday.append(((op-cl)/op))
    else:
        op_minus_cl_friday.append(0)
    if day == 1:
        op_minus_cl_monday.append(((op-cl)/op))
    else:
        op_minus_cl_monday.append(0)
    y += 1
    
df['Friday Return'] = op_minus_cl_friday
df['Monday Return'] = op_minus_cl_monday


correlation1=[]
correlation2=[]
z=0
while z < 7561:
    Friday_Return = df.iloc[z, df.columns.get_loc('Friday Return')]
    if Friday_Return != 0:
        correlation1.append(Friday_Return)
    z  += 1

i=0
while i < 7561:
    Monday_Return= df.iloc[i, df.columns.get_loc('Monday Return')]
    if Monday_Return != 0:
        correlation2.append(Monday_Return)
    i  += 1

correlation_matrix= {'friday': correlation1, 'monday': correlation2 }    

correlation_matrix

    
```
